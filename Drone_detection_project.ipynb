```python
import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn, FasterRCNN_ResNet50_FPN_Weights

# 1. Load model with default (COCO) weights
weights = FasterRCNN_ResNet50_FPN_Weights.DEFAULT
model = fasterrcnn_resnet50_fpn(weights=weights)
model.eval()

# 2. Prepare image (must be in 0-1 range, list of tensors)
# Here, we use random noise as a placeholder for an image [C, H, W]
input_image = [torch.rand(3, 800, 800)] 

# 3. Predict
with torch.no_grad():
    predictions = model(input_image)

print(predictions)
```
